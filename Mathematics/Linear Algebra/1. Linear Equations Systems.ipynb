## Introduction

### What is a Linear Equation?
A linear equation is an equation in which each term is either a constant or the product of a constant and a single variable. The general form of a linear equation in two variables is:

$$
ax + by = c
$$

where:
- $a$, $b$, and $c$ are constants.
- $x$ and $y$ are variables.

#### Examples:
1. $2x + 3y = 6$
2. $x - 4y = -5$
3. $y = 3x + 2$ (can be rearranged to $ -3x + y = 2$).

Linear equations can represent straight lines when graphed in two dimensions. The coefficients determine the slope and position of the line.

### Applications of Linear Equations
Linear equations are used extensively to model real-world situations, such as:
- Predicting trends in finance (e.g., cost and revenue models).
- Representing physical systems in physics (e.g., Ohm’s law: $V = IR$).
- Optimizing resources in operations research.

This section will lay the groundwork for understanding systems of linear equations, which involve multiple linear equations with shared variables. Let’s begin by exploring the theory behind linear systems.

## Theory of Linear Systems

### What is a System of Linear Equations?
A system of linear equations is a collection of two or more linear equations involving the same set of variables. The general form for a system of two equations with two variables is:

$$
\begin{align}
ax + by &= c \\
dx + ey &= f
\end{align}
$$

where:
- $a, b, c$ and $d, e, f$ are constants.
- $x$ and $y$ are the variables.

#### Example:
$$
\begin{align}
2x + 3y &= 6 \\
x - 4y &= -5
\end{align}
$$

### Types of Solutions
A system of linear equations can have:

1. **A Unique Solution**:
   - Occurs when the lines intersect at exactly one point.
   - Graphically, the two lines have different slopes.
   - Example: $x + y = 3$ and $x - y = 1$.

2. **No Solution**:
   - Occurs when the lines are parallel and never intersect.
   - Graphically, the lines have the same slope but different intercepts.
   - Example: $x + y = 3$ and $x + y = 5$.

3. **Infinitely Many Solutions**:
   - Occurs when the lines are coincident (i.e., they overlap completely).
   - Graphically, the two lines have the same slope and the same intercept.
   - Example: $x + y = 3$ and $2x + 2y = 6$.

### Representing Linear Systems
Linear systems can be represented in multiple ways:

1. **Equation Form**:
   - The standard representation using equations, e.g., $2x + 3y = 6$.

2. **Matrix Form**:
   - Using a coefficient matrix, variable vector, and a constant vector:
     $$
     \mathbf{A} \mathbf{x} = \mathbf{b}
     $$
     where:
     $$
     \mathbf{A} = \begin{bmatrix} a & b \\ d & e \end{bmatrix}, \quad \mathbf{x} = \begin{bmatrix} x \\ y \end{bmatrix}, \quad \mathbf{b} = \begin{bmatrix} c \\ f \end{bmatrix}
     $$

3. **Graphical Form**:
   - Plotting the equations on a graph to observe their intersection (if any).

### Graphical Interpretation
To understand the solutions of a system, it is helpful to plot the equations:

- **Intersection Point**: Unique solution.
- **Parallel Lines**: No solution.
- **Overlapping Lines**: Infinite solutions.

## Worked Examples

### Example 1: Solving a System with a Unique Solution
Consider the system:

$$
\begin{align}
2x + 3y &= 6 \\
x - 4y &= -5
\end{align}
$$

#### Step 1: Solve by Substitution
1. From the second equation, isolate $x$:
$$
x = 4y - 5
$$

2. Substitute $x = 4y - 5$ into the first equation:

$$
2(4y - 5) + 3y = 6
$$

3. Simplify:
$$
\begin{align}
8y - 10 + 3y &= 6 \\
11y &= 16 \\
y &= \frac{16}{11}
\end{align}
$$

4. Substitute $y = \frac{16}{11}$ back into $x = 4y - 5$:

$$
x = 4\left(\frac{16}{11}\right) - 5 = \frac{64}{11} - 5 = \frac{9}{11}
$$

5. Solution: 

$$
\left(x, y\right) = \left(\frac{9}{11}, \frac{16}{11}\right)
$$

### Example 2: Graphical Interpretation
Graph the two equations to confirm the solution.

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10, 10, 400)
y1 = (6 - 2 * x) / 3  # From 2x + 3y = 6
y2 = (x + 5) / 4      # From x - 4y = -5

plt.figure(figsize=(8, 6))
plt.plot(x, y1, label=r'$2x + 3y = 6$', color='blue')
plt.plot(x, y2, label=r'$x - 4y = -5$', color='orange')
plt.scatter([9/11], [16/11], color='red', zorder=5, label=r'Solution $(\frac{9}{11}, \frac{16}{11})$')
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.legend()
plt.title('Graphical Solution of the System')
plt.xlabel(r'$x$')
plt.ylabel(r'$y$')
plt.show()
```

### Example 3: A System with No Solution
Consider the system:
$$
\begin{align}
x + y &= 2 \\
x + y &= 5
\end{align}
$$

#### Analysis:
- Both equations have the same slope (parallel lines) but different intercepts.
- This system has no solution.

### Example 4: A System with Infinite Solutions
Consider the system:
$$
\begin{align}
x + y &= 3 \\

$$

#### Analysis:
- The second equation is a multiple of the first.
- The lines overlap entirely, leading to infinitely many solutions.
