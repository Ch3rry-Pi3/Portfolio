## <span style="color:#4682B4">**1. Overview of Pattern Recognition**</span> <a id="overview"></a>

Pattern recognition is a foundational concept in algorithmic trading. It involves identifying and interpreting recurring patterns, trends, and behaviors in financial data to predict future market movements. This chapter provides a comprehensive introduction to pattern recognition techniques and their significance in day trading.

Pattern recognition can be approached through multiple lenses, such as statistical analysis, machine learning, and technical analysis. From a statistical perspective, it entails extracting meaningful features from data and applying models to make predictions. Machine learning takes this further by automating pattern identification and decision-making based on historical data.

In day trading, technical analysis plays a critical role in recognising patterns in stock prices, volumes, and other market indicators. Traders analyse chart patterns and trends to anticipate price movements. These patterns are categorised into types such as reversal patterns, continuation patterns, and consolidation patterns.

Below is a Python implementation of fundamental equations and algorithms for pattern recognition in day trading. It demonstrates time series processing and prediction using Moving Averages.

```python
# Importing required libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### <span style="color:#4682B4">**4.1. Moving Average**</span> <a id="moving-average"></a>

A moving average is a widely used statistical tool in financial analysis and time series data processing. It smooths out short-term fluctuations by calculating the average of a data series over a specified window, providing a clearer view of underlying trends. This helps in identifying patterns, forecasting, and reducing noise in volatile datasets. Common applications include stock price analysis, signal processing, and performance tracking.

```python
def moving_average(data, window_size):
    """Calculate the moving average of the given time series data.

    The moving average is a widely used technical indicator in financial analysis,
    smoothing out short-term fluctuations and highlighting long-term trends in the data.

    Parameters
    ----------
    data : pandas.Series
        A time series of price data.
    
    window_size : int
        The size of the window over which the moving average is calculated.
        A larger window smooths the data more but introduces more lag.
    
    Returns
    -------
    pandas.Series
        A time series representing the calculated moving average over the specified window.

    Notes
    -----
    - The moving average is computed using a rolling mean.
    - NaN values are returned for the initial periods where the window size cannot be met.
    """
    
    return data.rolling(window=window_size).mean()
```

```python
# Step 1: Creating a sample time series data using pandas.Series
data = pd.Series([1, 2, 3, 4, 5])

# Step 2: Calculating the moving average of the data with a window size of 3
result = moving_average(data, window_size=3)

# Step 3: Printing the original data and the moving average results for comparison.
print("Original Data:\n")
print(data)
print("\nMoving Average (Window Size = 3):\n")
print(result)
```

### <span style="color:#4682B4">**4.2. Exponential Moving Average (EMA)**</span> <a id="ema"></a>

The Exponential Moving Average (EMA) is a time series analysis technique that assigns greater weight to recent data points, making it more responsive to recent changes compared to a simple moving average. This method is widely used in financial analysis to track trends and react quickly to price fluctuations. By exponentially decreasing the influence of older data, EMA helps identify momentum and signal potential reversals or continuations in market trends.

```python
def calculate_exponential_moving_average(data, span):
    """Calculate the Exponential Moving Average (EMA) of the given time series data.

    The EMA assigns greater weight to recent data points, making it more sensitive
    to recent price changes compared to a simple moving average.

    Parameters
    ----------
    data : pandas.Series
        A time series of price data.

    span : int
        The span (or lookback period) for the EMA calculation. A smaller span
        increases sensitivity to recent data.

    Returns
    -------
    pandas.Series
        A time series representing the calculated EMA.

    Notes
    -----
    - The EMA is computed using the pandas `ewm` method with `adjust=False`.
    - NaN values may appear at the beginning of the series where insufficient data exists for computation.
    """
    
    return data.ewm(span=span, adjust=False).mean()
```

```python
# Step 1: Creating a sample time series data using pandas.Series
data = pd.Series([1, 2, 3, 4, 5])

# Step 2: Calculating the Exponential Moving Average (EMA) of the data with a span of 2
result = calculate_exponential_moving_average(data, span=2)

# Step 3: Printing the original data and the EMA results for comparison
print("Original Data:\n")
print(data)
print("\nExponential Moving Average (Span = 2):\n")
print(result)
```

### <span style="color:#4682B4">**4.3. Buy and Sell Signals**</span> <a id="buy-sell-signals"></a>

The **Buy and Sell Signals** method generates trading signals based on the crossover of short-term and long-term moving averages. It identifies potential trading opportunities by comparing the trends in shorter and longer periods of the price data. 

- **Buy Signal:** Occurs when the short-term moving average crosses above the long-term moving average, indicating upward momentum.  
- **Sell Signal:** Occurs when the short-term moving average crosses below the long-term moving average, suggesting downward momentum.  

This approach is commonly used in technical analysis to guide trading decisions and identify entry or exit points in the market.

```python
def buy_sell_signals(data, short_window, long_window):
    """Generate buy and sell signals based on the crossover of short and long moving averages.

    This method identifies potential trading opportunities by comparing a shorter-term
    moving average with a longer-term moving average.

    Parameters
    ----------
    data : pandas.Series
        A time series of price data.
    
    short_window : int
        The window size for the short-term moving average.
    
    long_window : int
        The window size for the long-term moving average.

    Returns
    -------
    pandas.DataFrame
        A DataFrame containing:
        - 'short_mavg': The short-term moving average.
        - 'long_mavg': The long-term moving average.
        - 'signal': Buy (1) or Sell (0) signals based on the moving averages.
        - 'positions': Position changes (Buy: 1, Sell: -1, No change: 0).

    Notes
    -----
    - A buy signal is generated when the short-term moving average crosses above the long-term moving average.
    - A sell signal is generated when the short-term moving average crosses below the long-term moving average.
    """
    
    signals = pd.DataFrame(index=data.index)
    signals['short_mavg'] = moving_average(data, short_window)
    signals['long_mavg'] = moving_average(data, long_window)
    signals['signal'] = 0
    signals.loc[short_window:, 'signal'] = np.where(
        signals['short_mavg'][short_window:] > signals['long_mavg'][short_window:], 1, 0
    )
    signals['positions'] = signals['signal'].diff()
    return signals
```

```python
# Step 1: Creating a sample time series data using pandas.Series
data = pd.Series([1, 2, 3, 4, 5, 6])

# Step 2: Generating buy and sell signals based on short and long moving averages
signals = buy_sell_signals(data, short_window=2, long_window=3)

# Step 3: Printing the original data and the buy/sell signals for comparison
print("Original Data:\n")
print(data)
print("\nBuy/Sell Signals (Short Window = 2, Long Window = 3):\n")
print(signals)
```

### <span style="color:#4682B4">**4.4. Visualising Buy and Sell Signals**</span> <a id="plot-signals"></a>

The **Visualising Buy and Sell Signals** function creates a time series chart to display price data alongside short-term and long-term moving averages. It includes markers for buy and sell signals to visually highlight trading opportunities.

- **Buy Signal:** Represented with green upward triangles (`^`).  
- **Sell Signal:** Represented with red downward triangles (`v`).  

This visualisation provides an intuitive way to analyse trading strategies and assess the effectiveness of moving average crossovers in identifying trends. The plot includes a legend for easy interpretation.

```python
def plot_signals(signals):
    """Visualise price data along with buy and sell signals on a time series chart.

    This function creates a plot showing the short-term and long-term moving averages
    along with markers indicating buy and sell signals.

    Parameters
    ----------
    signals : pandas.DataFrame
        A DataFrame containing:
        - 'short_mavg': The short-term moving average.
        - 'long_mavg': The long-term moving average.
        - 'positions': Buy/Sell positions (1 for Buy, -1 for Sell).

    Returns
    -------
    None
        The function displays the plot directly.

    Notes
    -----
    - Buy signals are marked with green upward triangles.
    - Sell signals are marked with red downward triangles.
    - The plot includes a legend for clarity.
    """

    plt.figure(figsize=(14, 7))
    plt.plot(signals.index, signals['short_mavg'], label='Short Moving Average', alpha=0.75)
    plt.plot(signals.index, signals['long_mavg'], label='Long Moving Average', alpha=0.75)
    
    # Plot buy signals
    plt.plot(signals[signals['positions'] == 1].index, 
             signals['short_mavg'][signals['positions'] == 1], 
             '^', markersize=10, color='g', label='Buy Signal')
    
    # Plot sell signals
    plt.plot(signals[signals['positions'] == -1].index, 
             signals['short_mavg'][signals['positions'] == -1], 
             'v', markersize=10, color='r', label='Sell Signal')
    
    plt.title('Price and Buy/Sell Signals')
    plt.legend()
    plt.show()
```

```python
# Step 1: Generating a dummy time series for demonstration
np.random.seed(42)
price_data = pd.Series(np.random.randn(100).cumsum() + 100)

# Step 2: Defining parameters for moving averages
short_window = 20
long_window = 50

# Step 3: Generating buy and sell signals based on moving averages
signals = buy_sell_signals(price_data, short_window, long_window)

# Step 4: Plotting the price data along with the buy/sell signals
plot_signals(signals)
```
